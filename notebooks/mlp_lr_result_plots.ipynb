```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load your data
# Replace 'your_data.csv' with the path to your CSV file
df = pd.read_csv("./../docs/results/feature_model_learning_rate_comparison.csv")

# Step 2: Filter the data
# We're interested in the 'lr' for learning rate and 'final_val_accuracy'
plot_data = df[["lr", "final_val_accuracy"]]

# Step 3: Create the plot
plt.figure(figsize=(8, 8))
sns.boxplot(x="lr", y="final_val_accuracy", data=plot_data)
plt.title("Box Plot of Final Validation Accuracy for Each Learning Rate")
plt.xlabel("Learning Rate")
plt.ylabel("Final Validation Accuracy")


plt.savefig("./images/final_val_accuracy_boxplot.png")
```
