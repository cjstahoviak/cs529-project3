```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load your data
# Replace 'your_data.csv' with the path to your CSV file
df = pd.read_csv("./../docs/results/feature_model_batch_size.csv")

# Step 2: Filter the data
# We're interested in the 'lr' for learning rate and 'final_val_accuracy'
plot_data = df[["batch_size", "final_val_accuracy"]]

# Step 3: Create the plot
plt.figure(figsize=(8, 8))
sns.boxplot(x="batch_size", y="final_val_accuracy", data=plot_data)
plt.title("Box Plot of Final Validation Accuracy for Each Batch Size")
plt.xlabel("Batch Size")
plt.ylabel("Final Validation Accuracy")


plt.savefig("./images/final_val_accuracy_v_batch_size_boxplot.png")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load your data
# Replace 'your_data.csv' with the path to your CSV file
df = pd.read_csv("./../docs/results/feature_model_batch_size.csv")

# Step 2: Filter the data
# We're interested in the 'lr' for learning rate and 'training_time'
plot_data = df[["batch_size", "training_time"]]

# Step 3: Create the plot
plt.figure(figsize=(8, 8))
sns.boxplot(x="batch_size", y="training_time", data=plot_data)
plt.title("Box Plot of Final Validation Accuracy for Each Training Time")
plt.xlabel("Batch Size")
plt.ylabel("Final Validation Accuracy")


plt.savefig("./images/train_time_v_batch_size_boxplot.png")
```
